```python
#Code for fitting a straight line having values of x and y with the help of example: Change in temprature with time.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import scipy as sc

# defining variables: Time(t), Temprature(Temp), Slop(m), x_intercept(c)
t = np.arange(1,61)
Temp = np.random.randint(25,40,60)

# creating my own function
def line_fit(a,b):
    n = len(a)
    sum_a = np.sum(a)
    sum_b = np.sum(b)
    sum_ab = np.sum(a*b)
    sum_a2 = np.sum(a*a)
    # calculating the slope(m) and y intercept(c)
    m = (n * sum_ab - sum_a * sum_b) / (n * sum_a2 - sum_a**2)
    c = (sum_b - m * sum_a) / n
    return m, c

# calculateing slope and intercept using my function
m, c = line_fit(t,Temp)
print('Slope=',m)
print('Intercept=',c)

# now we will plot the obtaining results
# Plot the original data
plt.scatter(t, Temp, color='green', label='Data Points')

# Plot the fitted straight line
plt.plot(t, m * t + c, 'r', label='Fitted Line')

# Add labels and legend
plt.ylim(-100,100)
plt.xlabel('Time')
plt.ylabel('Temprature')
plt.legend()

# Show the plot
plt.show()
```
